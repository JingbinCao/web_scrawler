```python
import requests
from bs4 import BeautifulSoup

# 保存所有标题的列表
titles = []

# 爬取 1 到 20 页
for i in range(1, 21):
    url = f"https://www.t66y.com/thread0806.php?fid=7&search=&page={i}"
    # 使用 requests 库发送请求
    response = requests.get(url)
    # 使用 BeautifulSoup 库解析HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # 提取所有带有 "h3" 标签的元素
    h3_tags = soup.find_all('h3')

    # 从每个 "h3" 标签中提取标题
    for h3 in h3_tags:
        a_tag = h3.find('a')
        if a_tag:
            titles.append(a_tag.text.strip())

# 打印所有标题
for title in titles:
    print(title)
```
